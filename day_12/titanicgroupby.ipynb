```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic').loc[:,['age','sex', 'class', 'fare', 'survived']]
df
```

```python
import numpy as np
df1 = df.groupby('class').agg(np.std)
df1
```

```python
def min_max(x):
    return x.max()- x.min()

df2 = df.groupby('class').agg(min_max)
df2
```

```python
df3 = df.groupby('class').agg(['min','max'])
df3
```

```python
df4 = df.groupby('class').agg({
    "fare":['min','max'],
    "age":['mean']
})
df4
```

```python
df5 = df.groupby('class').filter(lambda x:x['age'].mean()<30)
df5
```
