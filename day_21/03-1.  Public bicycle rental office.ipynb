```python
import folium
import pandas as pd
import googlemaps
import numpy as np
from folium.plugins import MarkerCluster

gmaps_key= "AIzaSyBAgdunCaHubS6cDbxNrQyzoU5KReTwus0"
gmaps = googlemaps.Client(key=gmaps_key)

import json
geo_path = '../../data/python_data/02. skorea_municipalities_geo_simple.json'
geo_str=json.load(open(geo_path, encoding='UTF-8'))


bike = pd.read_excel('../../data/python_data/공공자전거 대여소 정보.xlsx',
                     engine ='openpyxl')

new_bike = bike[['구분','거치대수']]
new_bike = new_bike.set_index("구분")


mc = MarkerCluster()
mapping = folium.Map(location=[bike['위도'].mean(),bike['경도'].mean()],
                    zoom_start=11)

mapping.choropleth(geo_data=geo_str,
                      data = new_bike['거치대수'],
                      columns = [new_bike.index, new_bike['거치대수']],
                      fill_color='RdYlGn_r',
                      key_on = 'feature.id')

for n in bike.index:
    if bike['대여소명'][n] != 'null':
        mc.add_child(
            folium.Marker([bike['위도'][n], bike['경도'][n]]))
    
mapping.add_child(mc)
mapping
```
