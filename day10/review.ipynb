```python
import pandas as pd
encore= [{'name': 'A', 'birth': '1999-06-27', 'mid': 95, 'fin': 85},
              {'name': 'B', 'birth': '1997-06-27', 'mid': 85, 'fin': 80},
              {'name': 'C', 'birth': '1998-06-27', 'mid': 10, 'fin': 30},
              {'name': 'D', 'birth': '2000-06-27', 'mid': 73, 'fin': 90}]
df = pd.DataFrame(encore, columns = ['name', 'birth', 'mid', 'fin'])
df
```

```python
student = [{'name': 'A', 'birth': '1999-06-27', 'mid': 95, 'fin': 85},
              {'name': 'B', 'birth': '1997-06-27', 'mid': 85, 'fin': 80},
              {'name': 'C', 'birth': '1998-06-27', 'mid': 10, 'fin': 30},
              {'name': 'D', 'birth': '2000-06-27', 'mid': 73, 'fin': 90}]
df = pd.DataFrame(student, columns = ['name', 'birth', 'mid', 'fin'])
df
```

```python
df['tot'] = df['mid'] + df['fin']
df
```

```python
df['avg'] = df['tot']/2
df
```

```python
def grade(vec):
    if vec >= 90:
        return 'A'
    elif vec >= 80:
        return 'B'
    elif vec >= 70:
        return 'C'
    else:
        return  'F'
```

```python
df['grade'] = df['avg'].apply(lambda vec:grade(vec))
df
```

```python
def score(vec):
    if vec in ["A","B","C"]:
        return "합격"
    else:
        return "불합격"
```

```python
df['grade1'] = df['grade'].apply(lambda vec:score(vec))
df
```

```python
from datetime import datetime as dt


def get_age(vec):
    return  dt.today().year - vec + 1
    
```

```python
df['age'] = df['year'].apply(lambda vec:get_age(vec))
df
```
