## 자연어처리
- 자연어 : 일상 생활 속에서 사용하는 언어
- 자연어 처리 : 자연어 의미를 분석하여 컴퓨터가 처리할 수 있도록 하는 일
- 음성 인식, 내용 요약, 번역, 사용자의 감정 분석, 텍스트 분류 작어브 질의 응답 시스템, 챗봇과 같은 곳에서 사용
- 자연어 처리는 기계에게 인간의 언어를 이해시킨다는 점에서 인공지능에 있어서 가장 중요한 연구 분야이며, 최근 딥 러닝이 주목 받으면서, 인공지능이 제 4차산업의 중요 키워드로 떠오르는 중

### 분석 절차
- 요구사항 분석 > 텍스트 수집 > 텍스트 저장 및 전처리 > 텍스트 분석 > 텍스트 분석 서비스 > 산출물 관리 및 공유

### 데이터 전처리 순서
1. 토큰화 : 문자열 단어로 분리시키는 단계
2. 불용어 제거 : 전치사, 관사 등 너무 많이 등장하는 단어 등 문장이나 문서의 특징을 표현하는데 불필요한 단어를 삭제하는 단계
3. 어간 추출 : 단어의 기본 형태를 추출하는 단계
4. 문서 표현 : 주어진 문서나 문장을 하나의 벡터로 표현하는 단계, 단어들을 모두 인덱싱하고, 주어진 문서에 존재하는 단어의 빈도수를 사용하여 문장 표현

KoNLPy를 쓰면 파이썬으로 형태소 분석을 하거나 워드 클라우드를 그릴 수 있음 <br>
NLTK를 활용하면 문서 안의 내용을 빠르게 탐색하거나 요약 가능, 토필 모델링을 지원하는 패키지 Gensim을 사용하면 여러 분장이나 문서에 내재되어 있는 규칙, 토픽을 찾을 수 있음

## Naive Bayes Classifier의 이해 - 영문

```python
from nltk.tokenize import word_tokenize
import nltk
```

```python
# 지도학습, 비지도 학습 - 지도 학습의 예
# 훈련 세트 4개 문장에 라벨링 - 긍정, 부정
train = [('i like you','pos'),
        ('i hate you','neg'),
        ('you like me','neg'),
        ('i like her','pos')]
```

```python
for i in train:
    for j in word_tokenize(i[0]):
        print(j.lower())
```

```python
# 해당 단어들만 있으면 주어진 문장 만들 수 있음
all_words = set(word.lower() for sentence in train 
                            for word in word_tokenize(sentence[0]))
all_words
```

```python
# 말뭉치에 있는지 없는지로 표현
# i like you를 보자면, 위의 결과들 중에 me, hate, her은 사용되지 않았고 i like you를 사용하였음을 나타냄
t = [({word:(word in word_tokenize(x[0])) for word in all_words},x[1]) for x in train]
t
```

```python
# hate가 false일 때, pos:neg 1.7:1.0
classifier = nltk. NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
```

```python
test_sentence = 'i like MeRui'
test_sent_features = {word.lower():
                      (word in word_tokenize(test_sentence.lower()))
                                           for word in all_words}
test_sent_features
```

```python
classifier.classify(test_sent_features)
```

## Naive Bayes Classifier의 이해 - 영문

```python
from konlpy.tag import Twitter
```

```python
pos_tagger = Twitter()
```

```python
# 지도학습 - 훈련세트의 문장이 긍정, 부정 라벨링 되어 잇음
train = [('메리가 좋아','pos'),
        ('고양이도 좋아','pos'),
        ('난 수업이 지루해','neg'),
        ('메리는 이쁜 고양이야','pos'),
        ('난 마치고 메리랑 놀거야','pos')]
```

```python
all_words = set(word.lower() for sentence in train 
                            for word in word_tokenize(sentence[0]))
all_words
```

```python
# 말뭉치에 있는지 여부를 나타냄(5개의 훈련 세트)
t = [({word:(word in word_tokenize(x[0])) for word in all_words},x[1]) for x in train]
t
```

```python
test_sentence = '난 수업이 마치면 메리랑 놀거야'
```

```python
test_sent_features = {word.lower():
                      (word in word_tokenize(test_sentence.lower()))
                                           for word in all_words}
test_sent_features
```

```python
classifier.classify(test_sent_features)
```

```python
# stem=True: 원형, norm=True : 정규화(nornalize)
def tokenize(doc):
    return ['/'.join(t) for t in pos_tagger.pos(doc, norm=True, stem=True)]
```

```python
for row in train:
    print(tokenize(row[0]))
```

```python
from konlpy.tag import Twitter
twitter = Twitter()
print(twitter.pos('이것도 재미 있습니당ㅋㅋ'))
print('--------------------------------------------------------------------')
print(twitter.pos('이것도 재미 있습니당ㅋㅋ', norm=True))
print('--------------------------------------------------------------------')
print(twitter.pos('이것도 재미 있습니당ㅋㅋ', norm=True, stem=True))
```

```python
train_docs = [(tokenize(row[0]),row[1])for row in train]
train_docs
```

```python
tokens = [t for d in train_docs for t in d[0]]
tokens
```

```python
def term_exists(doc):
    return {word:(word in set(doc))for word in tokens}
```

```python
for d,c in train_docs:
    print(d,c)
```

```python
# 토큰라이징한 후 문장이 말뭉치에 있는지 확인
train_xy = [(term_exists(d), c) for d,c in train_docs]
train_xy
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_xy)
```

```python
test_sentence = [("난 수업이 마치면 메리랑 놀거야")]
```

```python
classifier.show_most_informative_features()
```

```python
test_docs = pos_tagger.pos(test_sentence[0])
test_docs
```

```python
test_sent_features = {word: (word in tokens) for word in test_docs}
test_sent_features
```

```python
classifier.classify(test_sent_features)
```
